```python
from dotenv import load_dotenv
load_dotenv( override=True)
```

```python
from langsmith import Client
client = Client()
prompt = client.pull_prompt("power_ranger")
```

copied power ranger sdk code from lansmith 



```python
prompt
```

```python
hydrated_prompt = prompt.invoke({"question": "Which type of power ranger you are ?"})
hydrated_prompt
```

```python
from openai import OpenAI
from langsmith.client import convert_prompt_to_openai_format

openai_client = OpenAI()

# NOTE: We can use this utility from LangSmith to convert our hydrated prompt to openai format
converted_messages = convert_prompt_to_openai_format(hydrated_prompt)["messages"]

openai_client.chat.completions.create(
        model="gpt-4o-mini",
        messages=converted_messages,
    )
```
